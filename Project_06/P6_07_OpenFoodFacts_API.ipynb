```python
save_path = './openfoodfacts_champagne/'
os.makedirs(save_path, exist_ok=True)
```

```python
import requests
import pandas as pd
import json
```

```python
# OpenFoodFacts API URL
url = "https://world.openfoodfacts.org/cgi/search.pl"
# Query parameters with more precise filters
params = {
    "action": "process",
    "tagtype_0": "categories",
    "tag_contains_0": "contains",
    "tag_0": "champagnes",  # Specific category for champagnes
    "tagtype_1": "categories",
    "tag_contains_1": "contains",
    "tag_1": "sparkling-wines",  # Additional category for sparkling wines
    "tagtype_2": "packaging",
    "tag_contains_2": "contains",
    "tag_2": "bottle",  # Bottle filter
    "sort_by": "popularity",
    "page_size": 10,
    "json": 1
}
```

```python
# Request the API
response = requests.get(url, params=params)
data = response.json()
```

```python
# Check if products have been found
if 'products' in data and len(data['products']) > 0:
    # Prepare a list to store product data
    products_data = []

    # Extract the information requested for each product
    for product in data['products']:
        product_info = {
            'foodId': product.get('code', ''),
            'label': product.get('product_name', ''),
            'category': product.get('categories', ''),
            'foodContentsLabel': product.get('ingredients_text', ''),
            'image': product.get('image_url', '')
        }
        products_data.append(product_info)

    # Create a pandas DataFrame with collected data
    df = pd.DataFrame(products_data)

    # Save data in a CSV file
    df.to_csv(save_path + 'champagne_products.csv', index=False)

    print(f"Extraction complete ! {len(products_data)} champagne products have been recorded in champagne_products.csv")

    # Display extracted data for verification
    print("\nOverview of extracted data :")
    print(df.head())
else:
    print("No champagne products were found in the OpenFoodFacts database.")

```
